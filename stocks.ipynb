```python
import yfinance as yf
```

```python
nvda_df = yf.download('NVDA', start='2020-01-01', end='2023-12-31')
```

```python
nvda_df.head()
```

```python
import matplotlib.pyplot as plt

nvda_df[('Close', 'NVDA')].plot(figsize=(12, 6), title='NVDA Close Price (2020-2023)')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.show()
```
